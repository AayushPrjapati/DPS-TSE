# Joint Mixture-Guided Target Speech Extraction (DPS-TSE) Demo

This notebook demonstrates **Joint Mixture Guidance (V1)**, a training-free hybrid framework for Target Speech Extraction (TSE) developed during the Summer Research Internship (SRI) program at the Speech Research Laboratory (SRL), DAU.

It combines a discriminative model (**TFGridNet**) with a generative diffusion prior (**ArrayDPS**) to isolate and refine the target speaker's voice naturally.
