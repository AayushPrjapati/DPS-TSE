# Joint Mixture-Guided Target Speech Extraction (DPS-TSE) Demo

This notebook demonstrates **Joint Mixture Guidance (V1)**, a training-free hybrid framework for Target Speech Extraction (TSE) developed during the Summer Research Internship (SRI) program at the Speech Research Laboratory (SRL), DAU.

It combines a discriminative model (**TFGridNet**) with a generative diffusion prior (**ArrayDPS**) to isolate and refine the target speaker's voice naturally.

```python
#@title Upload Custom WAV Files
from google.colab import files
import os

print("👉 Upload Mixture File (Multi-speaker audio):")
uploaded_mix = files.upload()
if len(uploaded_mix) > 0:
    mixture_path = list(uploaded_mix.keys())[0]
    print(f"Success: Mixture path is set to {mixture_path}")
else:
    mixture_path = "test_samples/LIBRI2MIX_mixture.wav"
    print(f"No file uploaded. Using default test sample: {mixture_path}")

print("\n👉 Upload Enrollment File (Target speaker reference audio):")
uploaded_enroll = files.upload()
if len(uploaded_enroll) > 0:
    enrollment_path = list(uploaded_enroll.keys())[0]
    print(f"Success: Enrollment path is set to {enrollment_path}")
else:
    enrollment_path = "test_samples/LIBRI2MIX_s1.wav"
    print(f"No file uploaded. Using default test sample: {enrollment_path}")
```
